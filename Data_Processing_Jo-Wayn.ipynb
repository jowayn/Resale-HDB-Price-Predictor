```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

pipeline = ColumnTransformer([
("oe", OrdinalEncoder(), ["month","flat_type","storey_range"]),
("oh", OneHotEncoder(), ["town", "flat_model"]),
], remainder='passthrough')
```

```python
X_pre = trimmed_data.drop(["resale_price"], axis=1)
```

```python
X = pipeline.fit_transform(X_pre)
y = trimmed_data["resale_price"]
```

```python
# Print the mappings from each column
for i, col in enumerate(["month","flat_type","storey_range","town", "flat_model"]):
    if col in pipeline.transformers_[0][2]:
        idx = pipeline.transformers_[0][2].index(col)
        mapping = dict(zip(range(len(pipeline.named_transformers_["oe"].categories_[idx])),
                           pipeline.named_transformers_["oe"].categories_[idx]))
    elif col in pipeline.transformers_[1][2]:
        idx = pipeline.transformers_[1][2].index(col)
        mapping = dict(zip(range(len(pipeline.named_transformers_["oh"].categories_[idx])),
                           pipeline.named_transformers_["oh"].categories_[idx]))
    else:
        mapping = {}
    print(f"{col}: {mapping}")
```

```python
# Extract the one-hot encoder column names and mappings
oh_cols = pipeline.transformers_[1][2]
oh_mappings = pipeline.named_transformers_["oh"].categories_

oe_names = ['month', 'flat_type', 'storey_range']

# Create the one-hot encoder column names
oh_names = []
for col, cats in zip(oh_cols, oh_mappings):
    oh_names.extend([f"{col}_{cat}" for cat in cats])

# Get the remaining column names
remaining_cols = [col for col in X_pre.columns if col not in oe_names and col not in oh_cols]

# Combine all the column names
column_names = oe_names + oh_names + remaining_cols
```

Linear Regression

```python
# Perform linear regression using train data and evaluate using test data:

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(X_train, y_train)

print("R-sq of model: ", lr.score(X_train, y_train))
pred = lr.predict(X_test)
print("RMSE: %.2f" % mean_squared_error(y_test, pred, squared=False))
print("R-sq of predictions: ", r2_score(y_test, pred))
```

KNN Regression

```python
# Perform K Nearest Neighbours Regression without hyperparameter tuning, using training data and evaluate using test data:

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

knn = KNeighborsRegressor(n_neighbors=5) #KNN default n_neighbours=5
knn.fit(X_train, y_train)

print("R-sq of model: ", knn.score(X_train, y_train))
pred = knn.predict(X_test)
print("RMSE: %.2f" % mean_squared_error(y_test, pred, squared=False))
print("R-sq of predictions: ", r2_score(y_test, pred))
```

```python
# Perform K Nearest Neighbours Regression with hyperparameter tuning, using training data and evaluate using test data:

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import make_scorer, mean_squared_error, r2_score

# Define parameter grid for KNN regression
param_grid = {'n_neighbors': [6, 7, 8, 9],
              'weights': ['uniform', 'distance'],
              'p': [1, 2],
              'algorithm': ['ball_tree', 'kd_tree']} #exclude 'brute' algorithm as it is prone to overfitting and suits smaller datasets well

# Create KNN regression model
knn = KNeighborsRegressor()

# Create scoring functions for Grid Search to decide best parameters
scoring = {'r2': 'r2', 'rmse': make_scorer(mean_squared_error, squared=False, greater_is_better=False)}

# Perform grid search
grid = GridSearchCV(knn, param_grid, scoring=scoring, refit='r2', cv=5)
grid.fit(X_train, y_train)

# Print best parameters and scores
print("Best parameters: ", grid.best_params_)
print("R-sq of model: ", grid.score(X_train, y_train))
pred = grid.predict(X_test)
print("RMSE: %.2f" % mean_squared_error(y_test, pred, squared=False))
print("R-sq of predictions: ", r2_score(y_test, pred))
```

```python
# Perform Decision Tree Regression with hyperparameter tuning, using training data and evaluate using test data:

from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import make_scorer, mean_squared_error, r2_score

# Define parameter grid for decision tree regression
param_grid = {'criterion': ['squared_error', 'friedman_mse'],
              'max_depth': [None, 5, 10],
              'min_samples_leaf': [2, 4, 6]
              }

# Create decision tree regression model
dtr = DecisionTreeRegressor(random_state=0)

# Create scoring functions for Grid Search to decide best parameters
scoring = {'r2': 'r2', 'rmse': make_scorer(mean_squared_error, squared=False, greater_is_better=False)}

# Perform grid search
grid = GridSearchCV(dtr, param_grid, scoring=scoring, refit='r2', cv=5)
grid.fit(X_train, y_train)

# Print best parameters and scores
print("Best parameters: ", grid.best_params_)
print("R-sq of model: ", grid.score(X_train, y_train))
pred = grid.predict(X_test)
print("RMSE: %.2f" % mean_squared_error(y_test, pred, squared=False))
print("R-sq of predictions: ", r2_score(y_test, pred))
```

```python
# Perform Random Forest Regression with hyperparameter tuning, using training data and evaluate using test data:

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer, mean_squared_error, r2_score

# Define parameter grid for random forest regression
# Note: setting max_depth to None will not limit the growth of each tree in random forest, hence set a limit to prevent overfitting
param_grid = {'n_estimators': [400, 500, 600],
              'max_depth': [5, 10, 15]
              } 

# Create random forest regression model
rf = RandomForestRegressor() 

# Create scoring functions for Grid Search to decide best parameters
scoring = {'r2': 'r2', 'rmse': make_scorer(mean_squared_error, squared=False, greater_is_better=False)}

# Perform grid search
grid = GridSearchCV(rf, param_grid, scoring=scoring, refit='r2', cv=5)
grid.fit(X_train, y_train)

# Print best parameters and scores
print("Best parameters: ", grid.best_params_)
print("R-sq of model: ", grid.score(X_train, y_train))
pred = grid.predict(X_test)
print("RMSE: %.2f" % mean_squared_error(y_test, pred, squared=False))
print("R-sq of predictions: ", r2_score(y_test, pred))
```
